```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/customer_segmentation_dataset.csv')
```

```python
df.info()
```

```python
df.drop('Segment',axis=1,inplace=True)
```

```python
sns.pairplot(df)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df=ss.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
wcss_li=[]
for i in range(2,10):
  km=KMeans(n_clusters=i)
  km.fit_predict(df)
  wcss_li.append(km.inertia_)
```

```python
plt.plot(range(2,10),wcss_li)
plt.title('Elbow Method For finding best k values for KMeans Clustering')
plt.xlabel('Number of Cluster')
plt.ylabel('WCSS Score')
plt.show()
```

```python
km=KMeans(n_clusters=3)
clusters=km.fit_predict(df)
```

```python
df=pd.DataFrame(df,columns=['AOV','Total Spent'])
```

```python
df['Clusters']=clusters
```

```python
sns.scatterplot(data=df,x='AOV',y='Total Spent',hue='Clusters')
plt.title('Cluster Analysis')
```
